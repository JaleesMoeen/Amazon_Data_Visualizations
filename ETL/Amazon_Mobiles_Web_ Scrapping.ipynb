```python
#defintion to scrap the web data
def scrape_amazon_product(url):
    # Send an HTTP request to the specified URL
    response = requests.get(url)
    
    if response.status_code == 200:
        # Save the entire HTML content to a variable
        html_content = response.text

        # Parse the HTML content of the page
        soup = BeautifulSoup(html_content, 'html.parser')
        
        #***************************   Products Details   *****************************
        
        # Extract information based on HTML structure
        product_title_element = soup.find('span', {'id': 'productTitle'})
           
        # price_element = soup.find('span', {'id': 'priceblock_ourprice'})
        price_element = soup.find('span', {'class': 'a-offscreen'})
        
        #Extract the start-rating based on HTML structure 
        #star_rating_element  = soup.find('span', {'id': 'acrPopover'})
        star_rating_element  = soup.find('span', {'class': 'a-icon-alt'})
        
        #Extract the No of 'Global Rating' based on HTML structure
        number_of_global_ratings = soup.find('div', {'class': 'a-row a-spacing-medium averageStarRatingNumerical'})

        # Extract the Discount Price from HTML structure
        #discount_element = soup.find( 'span ', {'class': "a-price a-text-price a-size-base"})
        
        # Extract model number from the product title or any other relevant element
        model_number = extract_model_number(product_title_element.text.strip())
        
        
        
        #***************************   Model Year  *****************************
        
#         year = soup.find_all('td',class_='a-size-base prodDetAttrValue')
#         for i in range(len(year)):
# #             year[i].text.strip()[1:]
# #         model_year = int(year[5].text.strip()[1:])

#         # Assuming year is a list of td elements
#         year = soup.find_all('td', class_='a-size-base prodDetAttrValue')

#         # Extract the text from each td element and store it in a list
#         year_values = [elem.text.strip()[1:] for elem in year]
       
#         #Now, check if there are at least 6 elements in the list
#         if len(year_values) >= 6:
#             # Access the 6th element and convert it to an integer
#             model_year = int(year_values[5])
#             print("Model Year:", model_year)
#         else:
#             print("Not enough elements for Model Year")

         # Find the div containing the Model Year
#         model_year_div = soup.find('h4', text='Model Year')

#         if model_year_div:
#             model_year_value = model_year_div.find_next('div', class_='a-section').text.strip()
#             #return model_year_value
#         else:
#             return None
#         # Find the div containing the Model Year
#         model_year_div = soup.find('h4', text='Model Year')

#         if model_year_div:
#             model_year_value = model_year_div.find_next('div', {'class': 'a-section'}).text.strip()
#             print("Model Year:", model_year_value)
#             print(model_year_div)
#         else:
#             print("Model Year not found.")

#         # Find the div containing the Model Year
#         model_year_div = soup.find('div', {'class': 'a-section'  }, 'h4', text='Year')

#         if model_year_div:
#             # Extract the Model Year value
#             model_year_value = model_year_div.find_next('div', {'class': 'a-section'}).text.strip()
#             print("Model Year:", model_year_value)
#         else:
#             print("Model Year not found.")
            
#         # Find the row containing "Model year"
#         model_year_row = soup.find('h4', text='Model Year')
#         if model_year_row:
#             # Extract the model year value
#             model_year_value = model_year_row.find('div', class_='a-section').text.strip()
#         else:
#             print("Model year not found.")
        
        
        
        #***************************   Customer Stars Ratings   *****************************
        
        # Find the table using the specified class
        customer_stars_percentages_table = soup.find('table', class_='a-normal a-align-center a-spacing-base')
        if customer_stars_percentages_table:
            # Initialize a dictionary to store the data
            customer_stars_percentages_table_data = {}
            # Loop through each row in the table
            for row in customer_stars_percentages_table.select('tr.a-histogram-row'):
                # Extract the text content of the cells in the row
                cells = row.select('td')
                if len(cells) == 3:
                    star_label = cells[0].text.strip()
                    percentage = cells[2].text.strip()
                    customer_stars_percentages_table_data[star_label] = percentage
                    
                    
                    
                    
        #***************************   Features Stars Ratings    *****************************

#         # Find the div containing attributes and ratings
#         attributes_div = soup.find('div', {'id': 'cr-summarization-attribute-attr-touch-screen'})
#         #console.log(attributes_div)
#         features_and_ratings = {}
#         if attributes_div:
#             # Initialize a dictionary to store features and ratings
#             touch_screen_text = attributes_div.find('span', {class_:'a-size-base a-color-base'}).text
#             print(touch_screen_text)

#             # Loop through each attribute div
#             for attribute_div in attributes_div.find_all('div', {class_: 'a-section a-spacing-none'}):
#                     # Extract feature name
#                 feature_name = attribute_div.find('span', {class_: 'a-size-base a-color-base'}).text.strip()
#         #                 print (feature_name)
#         #                 break

#                     # Extract rating value
#                 rating_value = attribute_div.find('span', {class_: 'a-size-base a-color-tertiary'}).text.strip()

#                     # Add feature and rating to the dictionary
#                 features_and_ratings[feature_name] = rating_value
#                # print(features_and_ratings)
#         else:
#             print("Attributes div not found.")

                    
                    

        
        
        #***************************   Dictionaries to Store All Data    *****************************
        
        # Create a dictionary to represent the data for this product
        product_data = {
            #'ModelNumber': model_number,
            'URL': url,
            'product_title': product_title_element.text.strip() if product_title_element else None,
            #'model_year' : model_year, # if model_year_value else None,
            'price': price_element.text.strip() if price_element else None,
            'star_ratings': star_rating_element.text.strip() if star_rating_element else None,
            'number_of_global_ratings' : number_of_global_ratings.text.strip() if number_of_global_ratings else None,
            #'Discount' : discount_element.text.strip() if discount_element else None
            #'HTMLContent': html_content
        }
        
        # Add customer reviews rating percentages to the product_data dictionary
        product_data['customer_stars_percentages'] = customer_stars_percentages_table_data
        
        # Add features stars ratings to the product_data dictionary
        #product_data['features_and_ratings'] = features_and_ratings
        
        
        return product_data
    else:
        print(f"Failed to fetch data from {url}. Status Code: {response.status_code}")
        return None
    
    
    
    
def extract_model_number(product_title):
    # Implement your logic to extract the model number from the product title
    # This is just an example, you may need to customize it based on your data
    words = product_title.split()
    for word in words:
        if word.startswith('Model'):
            return word.replace('Model', '')
    return None

```

# Scrap and Store the Iphone Details

```python

# Dictionary to store scraped data organized by product title
iphone_scraped_data = {}

# Loop through each URL and scrape data
for url in iphone_details_urls:
    data = scrape_amazon_product(url)
    if data:
        product_title = data.get('product_title')            
        if product_title:
          iphone_scraped_data[product_title] = data
        

# Save the data to a JSON fil
json_file_path = '../Scraped Data/scraped_iphone_data.json'

with open(json_file_path, 'w', encoding='utf-8') as jsonfile:
    json.dump(iphone_scraped_data, jsonfile, ensure_ascii=False, indent=4)

print(f"Scraped data saved to {json_file_path}")
```

```python

scraped_iphone_df = pd.DataFrame([dict(title=k,**v) for k, v in iphone_scraped_data.items()])

split=scraped_iphone_df['title'].str.split(',',expand=True)

split.columns=[f'Features_{i+1}' for i in range(split.shape[1])]

scraped_iphone_df = pd.concat([scraped_iphone_df,split],axis=1)

```

```python
# Create a DataFrame from the scraped data
#scraped_iphone_df = pd.DataFrame(iphone_scraped_data)

# Specify the path to save the CSV file
csv_file_path = "../Scraped Data/scraped_iphone_data.csv"

# Save the DataFrame to a CSV file
scraped_iphone_df.to_csv(csv_file_path, index=False)

#Display the DataFrame
scraped_iphone_df
```
